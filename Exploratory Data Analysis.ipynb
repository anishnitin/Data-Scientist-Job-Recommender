```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("inputdf.csv")
df = df.drop(['Unnamed: 0'], axis =1)
df['user_id'] = np.random.randint(1,5, size=len(df))
```

```python
df.describe()
```

## Univariate Analysis

```python
df_cat = df[['job_simp', 'job_state', 'Size','Sector','Revenue','Industry', 'Type of ownership']]
```

```python
df_cat1 = df[['Python', 'SQL','R','AWS','Spark',  'Excel']]
```

# Categorical Univariate Analysis - 2 Categories

```python
plt.rcParams['figure.figsize'] = [15, 6]
for i in df_cat1.columns:
    cat_num = df_cat[i].value_counts()
    print("graph for %s: total = %d" % (i, len(cat_num)))
    plt.pie(x=cat_num, autopct="%.1f%%",pctdistance=0.5)
    labels = df_cat1.index
    plt.legend(labels, loc="right")
    plt.show()
```

# Categorical Univariate Analysis - Multiple Categories

```python
plt.rcParams['figure.figsize'] = [15, 6]
for i in df_cat.columns:
    cat_num = df_cat[i].value_counts()
    print("graph for %s: total = %d" % (i, len(cat_num)))
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
    
```

# Continuous Univariate Analysis

```python
plt.rcParams['figure.figsize'] = [10, 15]
fig, axes = plt.subplots(nrows=3, ncols=1)
ax0, ax1, ax2 = axes.flatten()


ax0.hist(df['avg_salary'])
ax0.set_title('Average Salary Histogram')

ax1.hist(df['age'])
ax1.set_title('Age Histogram')

ax2.hist(df['Rating'])
ax2.set_title('Rating Histogram')

fig.tight_layout()
plt.show()
```

## Multivariate Analysis

```python
plt.rcParams['figure.figsize'] = [12, 8]
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df[['age','avg_salary','Rating','Python','SQL']].corr(),vmax=1, center=0, cmap=cmap,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot = True)
```
