```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis =1)
df = df.drop(['Headquarters'], axis =1)
df = df.drop(['Competitors'], axis =1)
```

#Salary_Estimate

```python
df = df[df['Salary Estimate'] != '-1'] #check if there is null in salary estimate column
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0]) #split text
minus_Kd = salary.apply(lambda x: x.replace('K','').replace('$','').replace(' Per Hour','')) #replace unnecessary characters
df['min_salary'] = minus_Kd.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = minus_Kd.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df.min_salary+df.max_salary)/2 #average salary calculation
```

#Location

```python
df['Location'] = df['Location'].apply(lambda x: x.replace('New Jersey','New Jersey, NJ').replace('Utah','Utah, UT').replace('California','California, CA').replace('Remote','Remote, REM').replace('Missouri','Missouri, MO').replace('Virginia','Virginia, VA').replace('United States','United States, USA'))
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1]) #extract state ID
```

#Other columns

```python
#replace -1 to unknowns in other string columns
df['Size'] = df['Size'].apply(lambda x: x.replace('-1','Unknown'))
df['Type of ownership'] = df['Type of ownership'].apply(lambda x: x.replace('-1','Unknown'))
df['Industry'] = df['Industry'].apply(lambda x: x.replace('-1','Unknown'))
df['Sector'] = df['Sector'].apply(lambda x: x.replace('-1','Unknown'))
df['Revenue'] = df['Revenue'].apply(lambda x: x.replace('-1','Unknown'))
```

#Job_Title

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    else:
        return 'na'
    
    
df['job_simp'] = df['Job Title'].apply(title_simplifier)
```

#Job Technology columns

```python
df['Python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['R'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df['Spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['AWS'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['Excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['SQL'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
```

#Founded

```python
df['age'] = df.Founded.apply(lambda x: x if x < 1 else 2020 - x)
```

# Final cleaned data

```python
df.to_csv("inputdf.csv")
```
